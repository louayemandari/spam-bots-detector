```python
from google.colab import drive
from os.path import join
import pandas as pd
import tensorflow as tf
from tensorflow import keras
import numpy as np

drive.mount('/content/drive')
fname1 = "spam_train.csv"
fname2 = "spam_test.csv"
dtrain = pd.read_csv(join(path_prefix, fname1))
dtest = pd.read_csv(join(path_prefix, fname2))
dtrain.head()
```

```python
dtrain.head()
```

#**profile_pic**: Whether the user has a profile picture or not.

#**nums/length_username**: The ratio of the number of numerical characters in the username to the length of the username.

#**fullname_words**: The full name of the user, split into word tokens.

#**nums/length_fullname**: The ratio of the number of numerical characters in the full name to the length of the full name.

#**name==username**: Whether the username and full name are literally the same.

#**description_length**: The length of the user's bio in characters.

#**external_url**: Whether the user has an external URL.

#**private**: Whether the user's profile is private or not.

#**posts**: The number of posts that the user has made.

#**followers**: The number of followers that the user has.

```python
dtrain.isna().sum()
```

```python
x = dtrain[['profile pic'	,'nums/length username',	'fullname words',	'nums/length fullname',	'name==username',	'description length'	,'external URL',	'private',	'#posts',
            '#followers','#follows']]
x
```

```python
y  = dtrain[['fake']]
y
```

```python
model = tf.keras.Sequential([
   keras.layers.Dense(640, activation='relu', input_shape=(11,)),

   keras.layers.Dense(124, activation='relu'),
   keras.layers.Dense(124, activation='relu'),

   keras.layers.Dense(124, activation='relu'),
   keras.layers.Dense(124, activation='relu'),


   keras.layers.Dense(124, activation='relu'),
   keras.layers.Dense(124, activation='relu'),




   keras.layers.Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy')

model.summary()
```

```python

# Create a callback that will stop training when the loss reaches 0.17
early_stopping = tf.keras.callbacks.EarlyStopping(monitor='loss', patience=100, min_delta=0.17, mode='min')

# Fit the model
model.fit(x, y, epochs=100, callbacks=[early_stopping])
```

```python
import tensorflow as tf
from sklearn.metrics import accuracy_score

# Get the test data
x_test = dtest[['profile pic', 'nums/length username', 'fullname words', 'nums/length fullname', 'name==username', 'description length', 'external URL', 'private', '#posts', '#followers', '#follows']]
y_test = dtest[['fake']]

# Predict the labels for the test data
y_predicted = model.predict(x_test)

# Convert the predicted probabilities to binary labels (0 or 1)
y_predicted_binary = tf.round(y_predicted)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_predicted_binary)

print('Accuracy:', accuracy)
```
